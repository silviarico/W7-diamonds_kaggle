```python
import pandas as pd
import numpy as np
```

```python
ej=pd.read_csv("sample_submission.csv",encoding = "ISO-8859-1")
train=pd.read_csv("train.csv",encoding = "ISO-8859-1")
test=pd.read_csv("test.csv",encoding = "ISO-8859-1")
train.info()
```

```python
train.isna().sum()
```

```python
import seaborn as sn
import matplotlib.pyplot as plt

corrMatrix = train.corr()
plt.figure(figsize=(10,5))
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
train.groupby(["cut"])["price"].max()
```

```python
def corte(x):
    if x =='Ideal':
        return 5
    if x =='Premium':
        return 4
    if x =='Very Good':
        return 3
    if x=='Good':
        return 2
    if x=='Fair':
        return 1
```

```python
train["corte"]=train.cut.apply(corte)
train.corte.value_counts()
```

```python
train.groupby(["clarity"])["price"].mean()
```

```python
def claridad(x):
    if x =='IF':
        return 10
    if x =='VVS1':
        return 8
    if x =='VVS2':
        return 7
    if x=='VS1':
        return 5
    if x=='VS2':
        return 4
    if x=='SI1':
        return 2
    if x=='SI2':
        return 1
    if x=='I1':
        return 0
    
```

```python
train["claridad"]=train.clarity.apply(claridad)
train.head()
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
train["colorn"] = le.fit_transform(train["color"])

```

```python
corrMatrix = train.corr()
plt.figure(figsize=(15,9))
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
train.drop(["x","y","z"],axis=1, inplace=True)
```

```python
corrMatrix = train.corr()
plt.figure(figsize=(15,9))
sn.heatmap(corrMatrix, annot=True)
plt.show()
```

```python
train.drop(["cut","color","clarity"],axis=1, inplace=True)
train.head()
```

```python
train.set_index('id', inplace=True)
```

```python
from sklearn.model_selection import train_test_split as tts
train_1=train.copy()
X=train_1.drop("price",axis=1)
y=train.price
X_train, X_test, y_train, y_test = tts(X,y, test_size=0.2)
```

```python
lr.fit(X_train,y_train)
```

```python
y_predict_train = lr.predict(X_train) 
y_predict = lr.predict(X_test) 
```

```python
"MSE Test", mean_squared_error(y_predict, y_test)

```

```python
"MSE Train", mean_squared_error(y_predict_train, y_train)

```

```python
train_2=train_1.copy()
```

```python
X=train_2.drop(["depth","corte"],axis=1)
y=train.price
```

```python

models={'ridge': Ridge(),
    'lasso': Lasso(),
    'sgd': SGDRegressor(),
    'knn': KNeighborsRegressor(),
    'grad': GradientBoostingRegressor(),
        'Random Forest':RandomForestRegressor(),}
```

```python
from sklearn import metrics
```

```python
for name, model in models.items():
    model.fit(X_train, y_train)


for name, model in models.items():
    y_pred = model.predict(X_test)
    print(f"------{name}------")
    print('MAE - ', metrics.mean_absolute_error(y_test, y_pred))
    print('MSE - ', metrics.mean_squared_error(y_test, y_pred))
    print('RMSE - ', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
    print('R2 - ', metrics.r2_score(y_test, y_pred))



```
